```python
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 200)
pd.set_option('display.max_columns', None)
```

```python
# Radi Bois
#trans_path = './releves/mouvement_rb_3.5-to-8.5.xls'
#trans_path = './releves/mouvement_rb_9.5-to-22.5.xls'
trans_path = './releves/mouvement_nr_10.5-to-22.5.xls'
cheques_path = './releves/nr_cheques-20.5.txt'

# Nezha
#trans_path = './releves/mouvement_nr_23.4-to-9.5.xls'
#cheques_path = './releves/nr_cheques.txt'
```

```python
# Read data from xls file
df_trans = pd.read_csv(trans_path, 
        delimiter="\t", skiprows=8, skipfooter=2, engine='python')
# Drop date_Valeur column and Debit column
df_trans = df_trans.drop(['date_Valeur', 'Debit'], axis=1)
df_trans.Libelle = df_trans.Libelle.astype(str)

print("Bank Transactions:", df_trans.shape[0])
df_trans = df_trans.sort_values('DateOperation')
df_trans.style.set_properties(subset=["Libelle"],**{'text-align': 'left'})
```

```python
df_cheques = pd.read_csv(cheques_path, dtype=str)
df_cheques.amount = df_cheques.amount.astype(int)
print("Cheques:", df_cheques.shape[0])
df_cheques
```

```python
# Lookup the df_cheques.chequeNo in the df_trans.Libelle column
# and return the corresponding Libelle
#result = df_trans[df_trans['Libelle'].str.contains('|'.join(df_cheques['chequeNo'].astype(str)))] #['Libelle']
#print(result)
df_trans['chequeNo'] = df_trans['Libelle'].str.extract(r'(\d*\.\d+|\d+)', expand=False) #.fillna(0).astype(int) 

#df_cheques['chequeNo'] = df_cheques['chequeNo'].astype(int)

# Concat df_cheques with df_trans
df_merged = df_cheques.merge(df_trans, 'outer').sort_values('DateOperation')
# Select rows where amount is NaN
print("VA Cheques:")
display(df_merged[df_merged['amount'].isnull()])

# Select rows where DateOperation is NaN
print("Unpaid cheques:")
display(df_merged[df_merged['DateOperation'].isnull()]["chequeNo"])

# Select rows where DateOperation is NaN
print("Paid cheques:", df_merged[df_merged['DateOperation'].notna()].shape[0])
display(df_merged[df_merged['DateOperation'].notna()])
```
